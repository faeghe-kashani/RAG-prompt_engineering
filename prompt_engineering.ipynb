**Create a pipeline with text-generation task**

```python
import torch
from transformers import AutoTokenizer,pipeline
model_id = "togethercomputer/Llama-2-7B-32K-Instruct"

tokenizer = AutoTokenizer.from_pretrained(model_id)
pipeline = pipeline(
    "text-generation",
    model=model_id,
    tokenizer=tokenizer,
    torch_dtype=torch.float16,
    device_map="auto",
)
```

**Some LLMs need prompts inside Tags. Test our pipeline without [INST]...[/INST] tags.**

```python
prompt_template = "Write a poem about cats"

sequences = pipeline(
    prompt_template,
    do_sample=True,
    top_k=10,
    temperature=0.7,
    top_p=0.95,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id,
    max_length=256,
)

for seq in sequences:
    print(f"Result: {seq['generated_text']}")
```

**Test our pipeline with proper Tags.**

```python
prompt_template = "[INST]\nWrite a poem about cats\n[/INST]\n\n"

sequences = pipeline(
    prompt_template,
    do_sample=True,
    top_k=10,
    temperature=0.7,
    top_p=0.95,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id,
    max_length=256,
)

for seq in sequences:
    print(f"Result: {seq['generated_text']}")
```

**Testing instruction and question**

```python
system_prompt = "Answer user questions carefully"
message = "Tell me about AI"
prompt_template = f'[INST] <<SYS>>\n{system_prompt}\n<</SYS>>\n\n'
prompt_template = prompt_template + f'{message} [/INST]\n\n'

sequences = pipeline(
    prompt_template,
    do_sample=True,
    top_k=10,
    temperature=0.7,
    top_p=0.95,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id,
    max_length=256,
)

for seq in sequences:
    print(f"Result: {seq['generated_text']}")
```
